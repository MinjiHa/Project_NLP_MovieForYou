```python
import pandas as pd
pd.set_option('display.max_row', 500)
pd.set_option('display.max_columns', 100)
import numpy as np
import sys
np.set_printoptions(threshold=sys.maxsize)
from konlpy.tag import Okt # 옛날 Twitter 클래스
from konlpy.tag import Kkma
import re

okt = Okt()
kkma = Kkma()
```

```python
# 다음리뷰

daum = pd.read_csv("daum_reviews.csv")
daum_ = daum.loc[:,['순번','영화명','reviews']]
daumwords = daum_['reviews'][3:50].sum()
```

```python
daumwords
```

```python
# 정규표현식
letters_only = re.sub('[^0-9가-힣\s\.\!\?]*', '', daumwords)
# letters_only = re.sub('[\n]*', '', letters_only) # \n 없애는 코드
# letters_only = re.sub('[\s]*', '', letters_only) # 띄어쓰기 다 붙이는 코드
letters_only
```

```python
# 스탑워즈 불러오기
df = pd.read_csv("stopwords726_3.txt")
stopwords = df.iloc[:,0].unique()
```

```python
okt = Okt()
# 정제된 리뷰로 형태소 테이블 만들기
sampledf = pd.DataFrame(okt.pos(letters_only, norm=True, stem=True))
# 명사, 형용사, 동사만 추출하기
sampledf = sampledf[(sampledf[1]=='Noun')|(sampledf[1]=='Adjective')|(sampledf[1]=='Verb')]
# 2글자 이상인 단어만 가져와서 리스트 만들기
spword = []
for word in sampledf[0]:
    if len(word)==2:
        if word not in (stopwords): # 스탑워즈 제외하고 출력하기
            spword.append(word)
# 단어별 등장횟수 세기
pd.DataFrame(spword)[0].unique()
# 조사 / 한글자단어 빠진 상태이기 때문에 최대한 정제된 상황임
# 이때 나오는 불용어 (예 : 어떻다) 등을 제거하기 위한 우리만의 불용어사전만들기 (플랫폼별로 하기) !
```
